```python
import numpy
from matplotlib import pyplot
%matplotlib inline

nx = 41
dx = 2/(nx-1)
nt = 20
dt = 0.025

u = numpy.ones(nx)
u[int(0.5/dx):int(1/dx+1)] = 2
un = numpy.ones(nx)

pyplot.plot(numpy.linspace(0,2,nx), u);
```

```python
for n in range(nt+1):
    un = u.copy()
    for i in range(1,nx):
        u[i] = un[i] - un[i] * dt/dx * (un[i] - un[i-1])

pyplot.plot(numpy.linspace(0,2,nx), u);
```
